```python
import torch, utils
from torch import nn

batch_size, num_steps = 32, 35
num_epochs, lr = 500, 2.0
train_iter, vocab = utils.load_time_machine_data(batch_size, num_steps)
num_inputs, num_hiddens, num_layers = len(vocab), 256, 2 # 通过超参数num_layers设置隐藏层的层数
device = utils.try_gpu()
lstm_layer = nn.LSTM(num_inputs, num_hiddens, num_layers)
net = utils.RNN(lstm_layer, len(vocab)).to(device)
utils.rnn_train(net, train_iter, vocab, lr, num_epochs, device)
```
