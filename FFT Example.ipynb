```python
import torchaudio
import torch
import matplotlib.pyplot as plt

# Load the audio file
waveform, sample_rate = torchaudio.load('/home/lindenbaum/Desktop/1.mp3')
print(sample_rate)
# Check if the waveform has 2 channels (stereo)
num_channels, num_samples = waveform.shape

# Apply FFT to each channel
fft_channel_1 = torch.fft.fft(waveform[0])
fft_channel_2 = torch.fft.fft(waveform[1])

# Compute the frequency axis
frequencies = torch.fft.fftfreq(num_samples, 1/sample_rate)

# Take the magnitude (absolute value) of the FFT
magnitude_channel_1 = torch.abs(fft_channel_1)
magnitude_channel_2 = torch.abs(fft_channel_2)
print(magnitude_channel_1.shape, magnitude_channel_2.shape, num_samples)
# Plot the FFT results
plt.figure(figsize=(14, 6))

# Plot for Channel 1
plt.subplot(1, 2, 1)
plt.plot(frequencies[:num_samples//2], magnitude_channel_1[:num_samples//2])
plt.title('Frequency Domain (Channel 1)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

# Plot for Channel 2
plt.subplot(1, 2, 2)
plt.plot(frequencies[:num_samples//2], magnitude_channel_2[:num_samples//2])
plt.title('Frequency Domain (Channel 2)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
